```python
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.image as mpi
import time
```

```python
camdir = '/home/ophiuchus/bvexcam'
```

```python
path = os.path.join(camdir,'latest_saved_image.bmp')
```

```python
fig = plt.figure(figsize = (15,20))
ax = fig.add_subplot(1,1,1)
while True:
    image = mpi.imread(path)
    ax.imshow(image)
    ax.set_title('Current time: '+str(time.ctime()))
    plt.show()
```
